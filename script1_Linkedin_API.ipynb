```python
## Importing Necessary Libraries and creating an instance of Linkedin Class
from linkedin_api import Linkedin
from constants import LINKEDIN_USER_EMAIL, LINKEDIN_USER_PASS

linkedin_api = Linkedin(LINKEDIN_USER_EMAIL, LINKEDIN_USER_PASS)
```

```python
## Assuming first and last names of any uset is given
First_name = 'Debraj'
Last_name = 'Chatterjee'
```

```python
## Creating the name (in lower case)
name = First_name + ' ' + Last_name
name = name.lower()
```

```python
## Searching for the profile with the keyword as the name
profiles = linkedin_api.search(params={'keywords': name}, limit=1)
# print(profiles)
```

```python
import pandas as pd
## Creating users' table
df = pd.DataFrame(profiles)
# df
```

```python
## Selecting top 10 links of linkedin profiles from the list
links_of_profiles = list(df['navigationUrl'][:10])
```

```python
## Creating an API Endpoint to get the profile data out of the links
import requests
from constants import API_KEY
import numpy as np
api_endpoint = 'https://nubela.co/proxycurl/api/v2/linkedin'
```

```python
## Scraping Data from the Profile links via API
final_data = []
for link in links_of_profiles:
    linkedin_profile_url = link
    api_key = API_KEY
    headers = {'Authorization': 'Bearer ' + api_key}
    ## getting the response from the API
    response = requests.get(api_endpoint,
                            params={'url': linkedin_profile_url,'skills': 'include'},
                            headers=headers)
    profile_data = response.json()
    # print(profile_data)
    result = {}
    result['link'] = link
    result['name'] = profile_data['full_name'] if profile_data['full_name'] else np.nan
    result['profile_pic_url'] = profile_data['profile_pic_url'] if profile_data['profile_pic_url'] else np.nan
    result['headline'] = profile_data['headline'] if profile_data['headline'] else np.nan
    result['about'] = profile_data['summary'] if profile_data['summary'] else np.nan
    result['location'] = profile_data['city'] + ', ' + profile_data['state'] if profile_data['city'] and profile_data['state'] else np.nan
    result['experiences'] = profile_data['experiences'][0] if profile_data['experiences'] else np.nan
    result['education'] = profile_data['education'][0] if profile_data['education'] else np.nan
    result['skills'] = profile_data['skills'] if profile_data['skills'] else np.nan
    final_data.append(result)
```

```python
## Creating the dataset and storing it in local
dataset = pd.DataFrame(final_data)
dataset.to_csv('linkedin_profiles.csv', index=False)
```
